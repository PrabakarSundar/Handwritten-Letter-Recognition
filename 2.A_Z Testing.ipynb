## <font color = "green"><center> Handwritten Letter Recognition

### Import packages

```python
import numpy as np
import glob
import cv2
from keras import backend as K
from keras.utils import np_utils
from numpy.testing import assert_allclose
from keras.models import model_from_json, load_model
from keras.callbacks import ModelCheckpoint
import matplotlib.pyplot as plt
import random
```

### Import dataset

```python
file_path = 'Dataset as Image/2.Test/**/*.png'
dataTest = []
labelTest = []  
cnumo = 0
print("-->Classifying Images. I't will take some time. Plese Wait.")
for filename in glob.iglob(file_path, recursive=True):
    FN = filename.split("2.Test")
    FN = FN[1]
    #print(FN[1])
    cnumo += 1
    image = cv2.imread(filename, 1)
    image = cv2.resize(image, (28, 28))
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    image = image.reshape(( 28, 28,1))
    #image=np.expand_dims(image,axis=2)
    dataTest.append (image)
    lab = ord(FN[1])-65
    labelTest.append(lab)
print("-->Classification Done")
print(cnumo)
dataTest = np.array(dataTest)
labelTest = np.array(labelTest)


X_test = dataTest.astype("float32")/255
print(X_test.shape)
Y_test = labelTest
print("-->Converting Data")
print("Number of samples to test: ",len(dataTest))
```

### Preprocessing

```python
X_test = X_test.reshape(X_test.shape[0], 28, 28, 1).astype('float32')
Y_test = np_utils.to_categorical(Y_test)

num_classes = Y_test.shape[1]
print("Total Classes:",num_classes)
```

```python
#Load saved model
filepath = "model-v2.h5"
new_model = load_model(filepath)
#Compile Model
new_model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
print("Done")
```

### Predict with the loaded model

```python
predict = new_model.predict(X_test, verbose=1)
```

### <font color = "green"> Display results

```python
right_pred = 0
wrong_pred = 0
for i in range(int(len(predict)/1)):
    j = i
    #j = random.randint(1,800)*5
    print(i+1,"Predicted Value:",chr(np.argmax(predict[j])+65),"||Confidence:",max(predict[i]))
    print("   Actual Value  :",chr(np.argmax(Y_test[j])+65))
    '''img_to_visualize = X_test[j] #use image from npy 
    img_to_visualize = img_to_visualize.astype("float32") 
    plt.imshow(np.squeeze(img_to_visualize), cmap='gray')
    plt.show()
    print("\n\n")'''
    if(chr(np.argmax(predict[j])+65) == chr(np.argmax(Y_test[j])+65)):
        right_pred += 1
    else:
        img_to_visualize = X_test[j] #use image from npy 
        img_to_visualize = img_to_visualize.astype("float32") 
        plt.imshow(np.squeeze(img_to_visualize), cmap='gray')
        plt.show()
        print("\n\n")
        wrong_pred += 1
```

### Performance Analysis

```python
print("Total Images:",len(Y_test))
print("Correct Predictions:",right_pred," ","Wrong Predictions",wrong_pred)
acc = right_pred/len(Y_test)
print("Accuracy:",acc)
```
